```python
import numpy as np
```

# Definições
Cria as definições das matrizes de rotação em torno dos eixos

```python
# Função de rotação em Z
def rotz(theta):
    return np.array([[np.cos(theta), -np.sin(theta), 0],
                     [np.sin(theta), np.cos(theta), 0],
                     [0, 0, 1]])

# Função de rotação em Y
def roty(theta):
    return np.array([[np.cos(theta), 0, np.sin(theta)],
                     [0, 1, 0],
                     [-np.sin(theta), 0, np.cos(theta)]])

# Função de rotação em X
def rotx(theta):
    return np.array([[1, 0, 0],
                     [0, np.cos(theta), -np.sin(theta)],
                     [0, np.sin(theta), np.cos(theta)]])
```

# Rotações
Configura com o usuário o conjunto de rotações.
Uma rotação é definida por:
- Eixo (X, Y ou Z)
- Ângulo (em graus)
- Tipo (pré ou pós)

Após configurar todas as rotações, cria a matriz de rotação final

```python
# Quantas rotações ao total
n = input("Quantas rotações? ")

# Garante que n é um inteiro
n = int(n)

# Arrays que gardam os valores de cada rotação
eixos = []
angulos = []

# Loop para ler os valores
for i in range(n):
    eixo = input("Eixo de rotação {} (x, y ou z): ".format(i))
    angulo = input("Ângulo de rotação {} (em graus): ".format(i))
    tipo = input("Tipo de rotação {} (1 - pós (eixo no corpo) ; 2 - pré (eixo inercial)): ".format(i))

    # Transforma o eixo em maiúsculo
    eixo = eixo.upper()

    # Garante que angulo é um float
    angulo = float(angulo)
    # Converte o ângulo para radianos
    angulo = np.radians(angulo)

    # Garante que tipo é um inteiro
    tipo = int(tipo)

    if tipo == 1:
        # Adiciona o eixo e o ângulo ao array
        eixos.append(eixo)
        angulos.append(angulo)
    elif tipo == 2:
        # Adiciona o eixo e o ângulo ao array
        eixos.insert(0, eixo)
        angulos.insert(0, angulo)

    print("Rotação {} adicionada.".format(i))
    print("-------------------------")
```

```python
# Cria um array para as matrizes de rotação
matrizes = []

# Loop para criar as matrizes de rotação
for i in range(n):
    if eixos[i] == "X":
        # Cria a matriz de rotação em X
        matriz = rotx(angulos[i])
    elif eixos[i] == "Y":
        # Cria a matriz de rotação em Y
        matriz = roty(angulos[i])
    elif eixos[i] == "Z":
        # Cria a matriz de rotação em Z
        matriz = rotz(angulos[i])

    print("Matriz {}:".format(i))
    print(matriz)
    # Adiciona a matriz ao array
    matrizes.append(matriz)

# Cria a matriz de rotação final
matriz_final = matrizes[0]

# Loop para multiplicar as matrizes
for i in range(1, n):
    # Multiplica a matriz final pela matriz atual
    matriz_final = np.matmul(matriz_final, matrizes[i])

# Imprime a matriz final
print("Matriz final:")
print(matriz_final)
```
