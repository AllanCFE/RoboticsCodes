```python
import matplotlib.pyplot as plt
import numpy as np
import math

# if using a jupyter notebook
%matplotlib inline    
```

```python
t = np.arange(0.0, 2*np.pi, 0.2)
x = 0.2 * np.sin(t)**3
y = 0.15 * np.cos(t) - 0.0725 * np.cos(2*t) - 0.03 * np.cos(3*t) - 0.015 * np.cos(4*t)
```

```python
print(x.__len__())
print(y.__len__())
```

```python
plt.plot(x, y)
plt.show()
```

```python
Rotation = [
    [9,9,1],
    [9,9,0],
    [0,0,9999999999999999999]
]

v4 = math.atan2(Rotation[1][2], Rotation[0][2])
v5 = math.atan2(Rotation[2][2], math.sqrt(Rotation[0][2]**2 + Rotation[1][2]**2))
v6 = math.atan2(-Rotation[2][1],-1* Rotation[2][0])

print(np.degrees(v4))
print(np.degrees(v5))
print(np.degrees(v6))
```

```python
Antigo = [0.2,0.15,0.0725,0.03,0.015]
NR = 0.35

Novos = [0,0,0,0,0]

for i in range(0,5):
    Novos[i] = Antigo[i] * NR/0.2

# Print com 3 casas decimais
print("Novos: ", ["%.3f" % i for i in Novos])
```
